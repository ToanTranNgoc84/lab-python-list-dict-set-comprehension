````python
def initialize_inventory(products):
    """
    The function initialize_inventory takes products as a parameter and initializes the inventory dictionary using a dictionary comprehension.
    """
    inventory = {product: int(input(f"Enter the quantity of {product}s available: ")) for product in products}
    return inventory

def get_customer_orders(products):
    """
    The function get_customer_orders prompts the user to enter the number of customer orders 
    and gathers the product names using a loop and user input. It uses comprehension.
    """
    customer_orders = {}
    while True:
        product_name = input("Enter the name of a product that a customer wants to order (t-shirt, mug, hat, book, keychain): ")
        if product_name in products:
            customer_orders[product_name] = 1
        else:
            print("Wo do not have this product. Please enter a valid product name (t-shirt, mug, hat, book, keychain).")
            continue
        add_another = input("Do you want to add another product? (yes/no): ")
        if add_another.lower() != 'yes':
            break
    return customer_orders


def calculate_total_price(customer_orders):
    """
    The function calculate_total_price calculates the total price of the customer order. It prompts the user to enter the price of each product and uses comprehension to calculate the total price.
    """
    total_price = sum(float(input(f"Enter the price of {product}: ")) * quantity for product, quantity in customer_orders.items())
    return total_price

def update_inventory(customer_orders, inventory):
    """
    The function update_inventory updates the inventory dictionary based on the customer orders. It uses comprehension to filter out products with a quantity of zero.
    """
    for product, quantity in customer_orders.items():
        inventory[product] -= quantity
    inventory = {product: quantity for product, quantity in inventory.items() if quantity > 0}
    return inventory

def calculate_order_statistics(customer_orders, inventory):
    """
    The function calculates the order statistics (total products ordered, and percentage of unique products ordered) and returns these values.
    """
    total_products_ordered = sum(customer_orders.values())
    total_inventory = sum(inventory.values())
    percentage_ordered = total_products_ordered / total_inventory * 100
    order_status = (total_products_ordered, percentage_ordered)
    return order_statistics

def print_order_statistics(order_statistics):
    """
    The function prints the order statistics.
    """
    total_products_ordered, percentage_ordered = order_statistics
    print("\nOrder Statistics:")
    print(f"Total Products Ordered: {total_products_ordered}")
    print(f"Percentage of Unique Products Ordered: {percentage_ordered:.1f}%")

def print_updated_inventory(inventory):
    """
    The function prints the updated inventory.
    """
    print("\nUpdated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")

# we define a list called products which contains the items
products = ["t-shirt", "mug", "hat", "book", "keychain"]

# calling the functions in the right order
inventory = initialize_inventory(products)
customer_orders = get_customer_orders(products)
total_price = calculate_total_price(customer_orders)
inventory = update_inventory(customer_orders, inventory)
order_statistics = calculate_order_statistics(customer_orders, inventory)

print_order_statistics(order_statistics)
print_updated_inventory(inventory)
print(f"\nTotal Price: {total_price:.1f}")
````
